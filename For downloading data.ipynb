```python
import csv
cursor2.execute('SELECT * from department_data')
with open('department_data.csv', 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow([i[0] for i in cursor2.description])
    csv_writer.writerows(cursor2)

cursor2.execute('SELECT * from employee_data')
with open('employee_data.csv', 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow([i[0] for i in cursor2.description])
    csv_writer.writerows(cursor2)

cursor2.execute('SELECT * from employee_details_data')
with open('employee_details_data.csv', 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow([i[0] for i in cursor2.description])
    csv_writer.writerows(cursor2)

cursor2.execute('SELECT * from model_evaluation_results')
with open('model_evaluation_results.csv', 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow([i[0] for i in cursor2.description])
    csv_writer.writerows(cursor2)

cursor2.execute('SELECT * from model_evaluation_table')
with open('model_evaluation_table.csv', 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow([i[0] for i in cursor2.description])
    csv_writer.writerows(cursor2)
print('Done')
```

```python
#joining the databases together
#change column name to dept_id
employee_data.rename(columns ={'department' : 'dept_id'}, inplace = True)
```
